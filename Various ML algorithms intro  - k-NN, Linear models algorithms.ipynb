```python
from sklearn.model_selection import train_test_split
```

```python
X,y = mglearn.datasets.make_forge()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
knn_predictions = knn.predict(X_test)
```

```python
knn_predictions
```

```python
y_test
```

```python
knn.score(X_test, y_test) # we got 100% accuracy
```

```python
# let us visualize actual boundary that predicts the result for k = [1,3,9]
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(1,3, figsize = (10,3))

for k, axe in zip([1,3,9], axes):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X, y)
    mglearn.plots.plot_2d_separator(knn, X, fill = True, eps = 0.5, ax = axe, alpha = .4)
    mglearn.discrete_scatter(X[:, 0], X[:,1],y, ax = axe)
    axe.set_title(f"{k} neighbours")
    axe.set_xlabel("feature 0")
    axe.set_ylabel("feature 1")
```

```python
#the more the neighbours, the less complex is the model
```

```python
#p.52 pdf
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, 
                                                    stratify = cancer.target, random_state = 1)
```

```python
#let us again use knn model with breast cancer dataset
knn_accuracy_test = []
knn_accuracy_training = []
for k in range(1,11):
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    accuracy = knn.score(X_test, y_test)
    knn_accuracy_test.append(accuracy)
    knn_accuracy_training.append(knn.score(X_train,y_train))
```

```python
plt.plot(range(1,11), knn_accuracy_training, label = 'training accuracy')
plt.plot(range(1,11), knn_accuracy_test, label = 'test accuracy')
plt.xlabel("number of neighbours for knn model")
plt.ylabel('accuracy')
plt.legend()
```

```python
from sklearn.neighbors import KNeighborsRegressor

X,y = mglearn.datasets.make_wave(n_samples=40)
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1)
```

```python
reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train, y_train)
```

```python
print(f"the score for KNN regression \
with 3 neighbors is : {reg.score(X_test,y_test)}. \n It is R^2 test")
```

```python
#Let us compare different number of neighbors for this regression problem
```

```python
fig, axes = plt.subplots(1,3, figsize = (15,4))
#create 1000 evenly spaced data points between -3,3
line = np.linspace(-3,3, 1000).reshape(-1,1)

for k, ax in zip([1,3,9], axes ):
    reg = KNeighborsRegressor(n_neighbors=k)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line)) #predictions for that line
    ax.plot(X_train, y_train, "^", c = mglearn.cm2(0), markersize = 8)
    ax.plot(X_test, y_test, "v", c = mglearn.cm2(1), markersize = 8)
    
    ax.set_title(f"{k} neighbors \n train score: {reg.score(X_train,y_train).round(2)}, \
    test score: {reg.score(X_test,y_test).round(2)}")
    
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")

axes[0].legend(["Model predictions", "Training data/target", "Test data/target"], loc = "best")
    
```

```python
# Linear models work great on high dimensional models
from sklearn.linear_model import LinearRegression 
```

```python
X,y = mglearn.datasets.load_extended_boston()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
print(f"Training  score: {lr.score(X_test, y_test).round(3)}")
print(f"Test score: {lr.score(X_train, y_train).round(3)}")
```

```python
#page 63 
```

```python
#another linear model is ridge regression
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge().fit(X_train,y_train)
print(f"Training score: {ridge.score(X_train,y_train)}")
print(f"Test score {ridge.score(X_test,y_test)}")
```

```python
#We may adjust the alpha to change a little bit the performance of model
ridge10 = Ridge(alpha = 10).fit(X_train,y_train)
print(f"Training score: {ridge10.score(X_train,y_train)}")
print(f"Test score {ridge10.score(X_test,y_test)}")
```

```python
ridge01 = Ridge(alpha = 0.1).fit(X_train,y_train)
print(f"Training score: {ridge01.score(X_train,y_train)}")
print(f"Test score {ridge01.score(X_test,y_test)}")
```

```python
# let us visualize coefficient magnitudes for ridge regression with different values of alpha
# and linear regression
plt.plot(ridge.coef_, "s", label = "Ridge alpha = 1")
plt.plot(ridge10.coef_, "v", label = "Ridge alpha = 10")
plt.plot(ridge01.coef_, "^", label = "Ridge alpha = 0.1")
plt.plot(lr.coef_, "o", label = "Linear regression")
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
plt.ylim(-25, 25)
plt.hlines(0,0, len(lr.coef_))
plt.legend()
```

```python
#The graph shows the magnitude of each feature - 106 - boston data
#it is visible which feature has the greatest impact on the target value (in this case price of house)
```

## Lasso - another alternative for linear regression 

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso().fit(X_train, y_train)
print(f"Test score: {lasso.score(X_train, y_train)}")
print(f"Training score: {lasso.score(X_test,y_test)}")
print(f"Number of features used: {np.sum(lasso.coef_ != 0)}")
```

```python
#in lasso regression the model tries to make as much coefficient as possible zero
#just to take into account those the most important features
```

```python
#the scores are realatively low, but we can manipulate some variables 
lasso001 = Lasso(alpha = 0.01, max_iter = 100000).fit(X_train, y_train)

print(f"Test score: {lasso001.score(X_train, y_train)}")
print(f"Training score: {lasso001.score(X_test,y_test)}")
print(f"Number of features used: {np.sum(lasso001.coef_ != 0)}")
```

```python
#as is visible, the quality improved and now 36 features are taken into account
```

```python
lasso00001 = Lasso(alpha = 0.0001, max_iter = 100000).fit(X_train, y_train)

print(f"Test score: {lasso00001.score(X_train, y_train)}")
print(f"Training score: {lasso00001.score(X_test,y_test)}")
print(f"Number of features used: {np.sum(lasso00001.coef_ != 0)}")
```

```python
#let us again plot the coefficients of the features
plt.plot(lasso.coef_, "s", label = "lasso alpha = 1")
plt.plot(lasso001.coef_, "v", label = "lasso alpha = 0.01")
plt.plot(lasso00001.coef_, "^", label = "lasso alpha = 0.0001")
plt.ylim(-25,25)
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
plt.plot(ridge01.coef_, "o", label = "Ridge alpha = 0.1")
plt.legend(ncol = 2, loc = (0, 1.05))
```

## Linear models for classification

### Most common linear classification algorithms:
####  - Logistic regression (yes - regression), and 
#### - linear support vector machines (linear SVMs)

```python
#Let us visualize the decision boundary 
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
```

```python
X,y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1)
```

```python
fig,axes = plt.subplots(1,2, figsize = (10,3))

for model, ax in zip ([LinearSVC(), LogisticRegression()], axes):
    clf = model.fit(X_train, y_train)
    mglearn.plots.plot_2d_separator(clf, X, fill = False, eps = 0.5,
                                   ax= ax, alpha= .7)
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax =ax)
    ax.set_title(f"{model}")
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1 ")
axes[0].legend()
```

## LogisticRegression model on the Brest Cancer dataset

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
                                                    stratify = cancer.target,
                                                   random_state = 42)
```

```python
logreg = LogisticRegression(max_iter= 10000).fit(X_train, y_train)
```

```python
print(f"Training = score: {logreg.score(X_train,y_train)}")
print(f"Test = score: {logreg.score(X_test,y_test)}")
```

```python
#Let us change the value of C
logreg100 = LogisticRegression(C = 100, max_iter= 10000).fit(X_train,y_train)
```

```python
print(f"Training = score: {logreg100.score(X_train,y_train)}")
print(f"Test = score: {logreg100.score(X_test,y_test)}")
```

```python
logreg001 = LogisticRegression(C = 0.01, max_iter= 10000).fit(X_train,y_train)
print(f"Training = score: {logreg001.score(X_train,y_train)}")
print(f"Test = score: {logreg001.score(X_test,y_test)}")
```

```python
#let us analyze the coefficient with those 3 different models

plt.plot(logreg.coef_.T, "o", label = "logreg C = 1")
plt.plot(logreg100.coef_.T, "v", label = "logreg C = 100")
plt.plot(logreg001.coef_.T, "v", label = "logreg C = 0.01")
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
plt.ylim(-5,5)
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation = 90)
plt.hlines(0,0, cancer.data.shape[1])
plt.legend()
```

```python
from sklearn.datasets import make_blobs 
X,y = make_blobs(random_state=42)
```

```python
mglearn.discrete_scatter(X[:,0],X[:,1], y)
plt.xlabel("Feature 0 ")
plt.ylabel("Feature 1")
plt.legend(["Class 0 ", "Class 1", "Class 2 "])
```

```python
#We can use linearSVC classifier on the dataset
linearsvm = LinearSVC().fit(X,y)
print(f"coefficient shape: {linearsvm.coef_.shape}")
print(f"Intercept shape: {linearsvm.intercept_.shape}")
```

```python
#Let us visualize the lines
mglearn.discrete_scatter(X[:,0], X[:,1], y)
line = np.linspace(-15,15)
for coef, intercept, color in zip(linearsvm.coef_, linearsvm.intercept_, 
                                  ['b','r','g']):
    plt.plot(line, -(line*coef[0]+ intercept)/coef[1], c = color)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1 ")
plt.legend(["Class 0", "Class 1", "Class 2", 
            "Line class 0 ", "Line class 1", "Line class 2"], loc = (1.01, 0.3))
mglearn.plots.plot_2d_classification(linearsvm, X, fill=True, alpha=.7)
```
